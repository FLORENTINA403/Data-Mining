**Perzgjedhja e numrit te Clusterave me Elbow Method**

```python
X = df.iloc[:, [4,5]].values
print(X[:5])

clustering_score = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'random', random_state = 42)
    kmeans.fit(X)
    clustering_score.append(kmeans.inertia_) 

plt.figure(figsize=(10,6))
plt.plot(range(1, 11), clustering_score)
plt.scatter(5,clustering_score[4], s = 200, c = 'red', marker='*')
plt.title('The Elbow Method')
plt.xlabel('No. of Clusters')
plt.ylabel('Clustering Score')
plt.show()
```

**Pozicionimi i centroide-ve dhe vizualizimi i te dhenave ne cluster-a**

```python
kmeans= KMeans(n_clusters = 5, random_state = 42)
kmeans.fit(X)
pred = kmeans.predict(X)
print(pred)

df['Cluster'] = pd.DataFrame(pred, columns=['cluster'] )
print('Number of data points in each cluster= \n', df['Cluster'].value_counts())
plt.figure(figsize=(14,6))
plt.scatter(X[pred == 0, 0], X[pred == 0, 1], c = 'brown', label = 'Cluster 0')
plt.scatter(X[pred == 1, 0], X[pred == 1, 1], c = 'green', label = 'Cluster 1')
plt.scatter(X[pred == 2, 0], X[pred == 2, 1], c = 'blue', label = 'Cluster 2')
plt.scatter(X[pred == 3, 0], X[pred == 3, 1], c = 'purple', label = 'Cluster 3')
plt.scatter(X[pred == 4, 0], X[pred == 4, 1], c = 'orange', label = 'Cluster 4')

plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:, 1],s = 300, c = 'red', label = 'Centroid', marker='*')
plt.xlabel('Shpenzimet')
plt.ylabel('tremujori')
plt.legend()
plt.title('Shpenzimet ne kuader te viteve')
```

**Percaktimi manual i numrit te clusterave dhe vizualizimi i te dhenave ne raportin viti-shuma me K-means**

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sb 
from mpl_toolkits.mplot3d import Axes3D 
from termcolor import colored as cl
from sklearn.preprocessing import StandardScaler 
from sklearn.cluster import KMeans 

plt.rcParams['figure.figsize'] = (20, 10)
sb.set_style('whitegrid')
df=pd.read_csv('shpenzimet_tot4.csv', encoding='ISO-8859-1')
del df["Drejtoria"]
del df["Kategoria_Ekonomike"]
df['Viti']=df['Viti'].astype(int)
df['Shuma'] = df['Shuma'].astype(int)

X = df.values
X = np.nan_to_num(X)
sc = StandardScaler()

cluster_data = sc.fit_transform(X)
clusters = 3
model = KMeans(init = 'k-means++', 
               n_clusters = clusters, 
               n_init = 12)
model.fit(X)
labels = model.labels_
df['cluster_num'] = labels
df.head()
df.groupby('cluster_num').mean()


sb.scatterplot(df['Viti'].astype(str), df['Shuma'].astype(int), 
               data = df, 
               s = 380, 
               alpha = 0.8, 
               hue = 'cluster_num', 
               palette = 'spring', 
               edgecolor = 'darkgrey')
plt.title('Shpenzimet nder Vite', 
          fontsize = 18)
plt.xlabel('Viti', 
           fontsize = 16)
plt.ylabel('SHpenzimet', 
           fontsize = 16)
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.legend(loc = 'upper left', fontsize = 14)
plt.show()
```

**Paraqitja grafike 3D e te dhenave ne raportin viti-shuma-tremujori**

```python
fig = plt.figure(1)
plt.clf()
ax = Axes3D(fig, 
            rect = [0, 0, .95, 1], 
            elev = 48, 
            azim = 134)

plt.cla()
ax.scatter(df['tremujori'], df['Viti'], df['Shuma'], 
           c = df['cluster_num'], 
           s = 200, 
           cmap = 'spring', 
           alpha = 0.5, 
           edgecolor = 'darkgrey')
ax.set_xlabel('tremujori', 
              fontsize = 16)
ax.set_ylabel('Viti', 
              fontsize = 16)
ax.set_zlabel('Shuma', 
              fontsize = 16)

plt.show()
```
